1. What is the difference between a function and a method in Python?



Ans:-
 Function -
1. Function is a block of code which performs a specific task and can be called again and again as per requirement in the code.
2. Function can be called independently.
3. Function can be used for small reusable piece of code.

```python
#Example
def func(name):
  return f"Hello My Name is {name}"

print(func("Shayan"))
```

Method -
1. A Method is a function that is associated with an object. It is defined inside a class and can be called by creating an instance of the class.
2. Method cannot be called independently but can be called by creating object.
3. Method can be used for operation related to object and class.

```python
#Example
class Hello:
  def func(self, name):
    return f"Hello My Name is {name}"

hello = Hello()
print(hello.func("Shayan"))
```

2. Explain the concept of function arguments and parameters in Python.

Ans:-
1. The variable which we pass inside function while defining a function is called arguments, they are used to pass some input to the function.
2. Function arguments are the values that you provide to the function while calling a function, these are used by function to perform some specific task.

```python
#Example
def func(name): #Here name is the argument
  return f"My Name is {name}"

print(func("Shayan"))
```

We can pass arguments to the function by different ways.


1. Default Arguments - You can provide default values for function parameters, if the argument is not passed during calling then default value is used.


```python
#Example
def sum_of_two_numbers(a, b = 8): #Here b is the default argument
  return a + b

print(sum_of_two_numbers(4))
```

2. Keyword Argument - Keyword argument is used to specify the parameter name explicitly. This allow you to pass them in any order.

```python
#Example
def product_of_two_numbers(a, b):
  return a + b

print(product_of_two_numbers(a = 9, b = 8))
```

3. Variable Length Arguments - These arguments are used to pass variable inputs to the function.

```python
#Example
def sum_of_any_numbers(*args):
  sum = 0
  for i in args:
    sum += i
  return sum

print(sum_of_any_numbers(2, 3, 4, 5, 6, 7))
```

4. Key Value Arguments - These are used to pass key value pairs to the function.

```python
#Example
def func(**kwargs):
  return kwargs

print(func(a = 10, b = 20, c = 30, d = 40, e = 50))
```

3. What are the different ways to define and call a function in Python?

Ans:- There are different ways to define and call function in python.

1. **Function with no argument and no return value** :- In this simply we define a function without passing any argument and then call a function without passing any input.

```python
#Example
def func(): # defining a function with no argument
  print("Hello World!")

print(func()) #calling a function with no argument
```

2. **Functions with input argument and a return value** - We can pass any number of arguments inside a function.

```python
def func(name): #defining a function with argument
  return f"Your name is {name}"

print(func("Shayan")) #Calling a function with argument
```

3. **Function with default argument** - We can pass a default value to the function by default argument.

```python
#Example
def sum_2(a, b = 9): #function with default argument
  return a + b

print(sum_2(4))
```

4. **Nested Function** - We can define function inside function.

```python
#Example
def outer_func(name):
  def inner_func(greeting):
    return f"{greeting} {name}"
  return inner_func("Hello")

print(outer_func("Shayan"))
```

5. **Function with variable length argument** - We can define function with passing mutiple arguments together.

```python
#Example
def func(*args):
  return args

print(func(1, 2, 3, 4, 5))
```

6. **Function with key value argument** - We can define function by passing key and value pair.

```python
#Example
def func(**kwargs):
  return kwargs

print(func(a = 100, b = 90))
```

7. **Recursive Function** - We can define a recursive function which calls itself.

```python
#Example
def factorial(n):
  if n <= 1:
    return 1
  else:
    return n * factorial(n-1)

print(factorial(5))
```

8. **Lambda Function** - We can define a lambda function which takes two inputs.

```python
greet = lambda name: f"Hello {name}"

print(greet("Shayan"))
```

9. **Function with no argument but a return value**

```python
def func():
  a = int(input("Enter First Number: "))
  b = int(input("Enter Second Value: "))
  return a + b
func()
```

7. What are the advantages of using generators over regular functions?

Ans:-
Advantages of generators over regular functions:-

1. Generator returns the value or result one by one while Regular function returns the value or result in one go.

2. Generator does not load the entire sequence into memory while regular functions load the entire sequence into memory.

3. Generator can be applied easily to the large sequence of data while Regular functions takes lot of execution time when given large sequence of data

4. Generator only executes the given yield statement and pauses the further execution while Regular function executes the result in one go and then exits from the function.

5. Generators are memory efficient because they do not store all values in the memory once while Regular functions are not memory efficient.

6. Generators generates the value one at a time making them suitabke for working with large datasets when memory usage is concern while Regular functions generate the result in a single shot.

```python
#Example of Regular Function
def cube_of_n_numbers(n):
  result = []
  for i in n:
    result.append(i**3)

  return result

print(cube_of_n_numbers([1, 2, 3, 4, 5]))
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans:
map() function - map() function executes a specified function for each item of an iterable and return the result as iterable map object. The item is sent to the function as parameter.

```python
#Example
l = [1, 2, 3, 4, 5]
result = list(map(lambda x : x**2, l))
print(result)
```

```python
#Example
from functools import reduce

l = [2, 1, 3, 4, 4, 5, 6]
result = reduce(lambda x, y : x+y, l)
print(result)
```

```python
#Example
s = ["pwskills", "shayan", "Ajay"]

result  = list(filter(lambda x: len(x) > 4, s))
print(result)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_Prime(n):
  if n <= 1:
    return False
  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False
  return True

for i in range(1, 201):
  if is_Prime(i):
    print(f"{i} is a prime number")
  else:
    print(f"{i} is not a prime number")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Iterator:
  def __init__(self, n):
    self.n = n
    self.a = 0
    self.b = 1
    self.count = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.count >= self.n:
      raise StopIteration
    self.count += 1
    result = self.a
    self.a, self.b = self.b, self.a + self.b
    return result


obj = Iterator(10)
for num in obj:
  print(num, end = " ")
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_two(exponent):
  for n in range(exponent+1):
    yield 2 ** n

for i in power_of_two(10):
  print(i, end = " ")
```
