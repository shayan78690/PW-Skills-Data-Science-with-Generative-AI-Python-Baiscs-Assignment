#Q.2 How does broadcasting work in NumPy?

Ans:- Broadcasting in NumPy is a way to perform arithmetic operations on arrays of different shapes and sizes. Instead of manually resizing or reshaping arrays to match each other, NumPy automatically adjusts them to make the shapes compatible for element-wise operations. This makes calculations more efficient and faster.

NumPy compares the shapes of arrays and follows a set of rules to automatically apply the operation:-

1. If the arrays have the same number of dimensions and each dimension is either the same size or one of them is 1, the operation can be broadcasted.

2. If the arrays have different numbers of dimensions, NumPy will add dimensions to the smaller array to match the size of the larger one, following the rules above.

```python
#Example
import numpy as np

arr1 = np.array([[1], [2], [3]])
arr2 = np.array([10, 20, 30, 40])

result = arr1 + arr2
print(result)
```

#Q.4 Explain the use of the groupby() method in Pandas?

Ans:- The groupby() method in Pandas is used to group data based on certain characteristics (like column values) and then apply some operation (such as sum, mean, count, etc.) to each group. It's like splitting your data into different categories, performing calculations on each category, and then combining the results.

```python
#Example
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob', 'Charlie'],
    'Department': ['HR', 'Finance', 'HR', 'Finance', 'HR', 'Finance'],
    'Salary': [5000, 6000, 5500, 6500, 5700, 6800]
}

df = pd.DataFrame(data)

grouped = df.groupby('Department')['Salary'].mean()

print(grouped)
```

#Q.17 Explain the difference between apply() and map() in Pandas.

Ans:- Difference Between apply() and map() in Pandas:

1. Functionality:
* apply(): Can be used on both Series and DataFrame. It applies a function along an axis (rows or columns) of a DataFrame or to all elements of a Series.
* map(): Used only on Series to apply a function element-wise. It’s more efficient for element-wise transformations.

2. Use Case:
* apply(): More versatile, can be used for complex operations like aggregations or transformations across rows or columns.
* map(): Primarily used for transforming or substituting values in a Series, typically with a dictionary or a function.

3. Performance:
* apply(): Generally slower compared to map() because it allows more flexibility.
* map(): Faster for simple element-wise operations.

```python
#Example
import pandas as pd

data = pd.Series([1, 2, 3, 4])

apply_result = data.apply(lambda x: x**2)

map_result = data.map(lambda x: x * 2)

print("Apply Result:\n", apply_result)
print("Map Result:\n", map_result)
```

#Q.18 What are some advanced features of NumPy?

Ans:- Advanced Features of NumPy:
1. Broadcasting: Allows NumPy to perform operations on arrays of different shapes, automatically adjusting them to match each other.

```python
#Example
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4])
result = arr1 + arr2  # Broadcasting arr2 to match arr1 shape
print(result)  # Output: [5 6 7]
```

```python
#Example
arr = np.array([1, 2, 3])
result = arr * 2  # No need for loops, faster computation
print(result)  # Output: [2 4 6]
```

```python
#Example
arr = np.array([1, 2, 3, 4, 5])
result = arr[arr > 2]  # Indexing with a condition
print(result)  # Output: [3 4 5]
```

```python
#Example
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result = np.dot(A, B)  # Matrix multiplication
print(result)  # Output: [[19 22] [43 50]]
```

5. Random Module: Provides powerful methods for generating random numbers, including random sampling, distributions, and permutations.

```python
#Example
random_arr = np.random.rand(3, 3)  # Generate random 3x3 matrix
print(random_arr)
```

#Q.19 How does Pandas simplify time series analysis?

Ans:-
1. DateTime Indexing: Pandas allows easy handling of time series data using DatetimeIndex. It enables efficient indexing, slicing, and resampling based on dates and times.

```python
#Example
import pandas as pd

# Creating a time series
dates = pd.date_range('20240101', periods=6)
data = pd.Series([10, 20, 30, 40, 50, 60], index=dates)

print(data)
```

2. Resampling: You can easily change the frequency of time series data (e.g., from daily to monthly or vice versa) using the resample() method.

```python
#Example
data_resampled = data.resample('3D').mean()  # Resample data by 3 days
print(data_resampled)
```

3. Time Shifting: Pandas supports shifting the data forward or backward in time using the shift() method, useful for computing time-based operations like differences between periods.

```python
#Example
shifted_data = data.shift(1)  # Shift data by 1 period
print(shifted_data)
```

4. Rolling Windows: The rolling() method allows you to calculate moving averages and other statistical operations over a window of time.

```python
#Example
rolling_mean = data.rolling(window=2).mean()  # Calculate 2-period rolling mean
print(rolling_mean)
```

5. Time Zone Handling: Pandas allows converting between time zones and handling time zone-aware datetime objects easily.

```python
#Example
data_with_timezone = data.tz_localize('UTC').tz_convert('US/Eastern')
print(data_with_timezone)
```

#Q.20 What is the role of a pivot table in Pandas?

Ans:- A pivot table in Pandas is used to summarize, aggregate, and reshape data. It allows you to transform data into a more meaningful format, where you can aggregate values based on one or more columns.

Key Features:
* Data Aggregation: It groups data based on specific columns and performs an aggregate function (e.g., sum, mean, count).
* Reshaping: It reshapes data, making it easier to analyze large datasets by transforming them into a table format.
* Flexibility: It allows you to create multi-level summaries by defining multiple columns for rows and columns.

```python
#Example
import pandas as pd

data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 30, 40, 50],
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']
}

df = pd.DataFrame(data)

pivot_table = pd.pivot_table(df, values='Value', index='Category', aggfunc='sum')
print(pivot_table)
```

#Q.21 Why is NumPy’s array slicing faster than Python’s list slicing?

Ans:-
1. Underlying Implementation:
* NumPy Arrays are implemented in C and use contiguous memory blocks. This allows NumPy to quickly access and manipulate large datasets with low-level optimizations.
* Python Lists, on the other hand, are implemented as dynamic arrays with additional overhead for managing metadata (e.g., size, type). This makes list operations slower compared to NumPy arrays.

2. No Copying in NumPy:
* When slicing a NumPy array, the result is a view on the original array (except in certain cases like when modifying the shape). This means NumPy only creates a reference to the data, not a full copy, making it faster.
* In Python lists, slicing generally creates a new copy of the sliced elements, which incurs additional memory and time overhead.

3. Optimized Vectorized Operations:
* NumPy can apply vectorized operations on entire slices or sub-arrays at once, thanks to its internal optimizations. This enables high performance when working with large data sets.
* Python lists do not have native support for vectorized operations, so operations on slices are typically slower.

```python
#Example
import numpy as np

# NumPy array slicing
np_array = np.array([1, 2, 3, 4, 5])
np_slice = np_array[1:4]  # Fast slicing (view on original array)

# Python list slicing
py_list = [1, 2, 3, 4, 5]
py_slice = py_list[1:4]  # Slower slicing (creates a new list)
```
